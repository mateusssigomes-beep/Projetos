```python
import sqlite3

conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

print('Conectado a obanco de dados do sqlite')
```

```python
cursor.execute("""DROP TABLE IF EXISTS [TB_VIDEO]""")
cursor.execute("""DROP TABLE IF EXISTS [TB_FILME]""")
cursor.execute("""DROP TABLE IF EXISTS [TB_USUARIO]""")


print('Tabelas Excluidas com sucesso')
```

```python
cursor.execute("""
CREATE TABLE TB_VIDEO (
    id INTEGER PRIMARY KEY,
    titulo VARCHAR NOT NULL,
    imdb_rating INT NOT NULL,
    visto BOOLEAN NOT NULL
    );
    """)

print('Tabela Video criada')


cursor.execute("""CREATE TABLE TB_FILME (
    id INTEGER PRIMARY KEY,
    diretor VARCHAR NOT NULL,
    ano INT NOT NULL
    );
    """)

print('Tabela Filme criada')

cursor.execute("""CREATE TABLE TB_USUARIO (
    id INTEGER PRIMARY KEY,
    nome VARCHAR NOT NULL,
    nivel_acesso VARCHAR NOT NULL,
    usuario VARCHAR NOT NULL,
    senha VARVHAR NOT NULL
    );
    """)

print('Tabela Usuário criada')
```
